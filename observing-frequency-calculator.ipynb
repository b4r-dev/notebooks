# Observing freqency calculator

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-darkgrid')
plt.style.use('seaborn-muted')
plt.rcParams['figure.figsize'] = 12, 4
plt.rcParams['figure.dpi'] = 150
```

## Constants

```python
xffts_lower = 0.1 # GHz
xffts_upper = 2.4 # GHz
```

## Plotter

```python
def plot(f_LO1=140.0, f_LO2=8.1, lines=None, ax=None):
    if f_LO2 == 8.1:    
        usb_lower = f_LO1 + f_LO2 - xffts_upper
        usb_upper = f_LO1 + f_LO2 - xffts_lower
        lsb_lower = f_LO1 - f_LO2 + xffts_lower
        lsb_upper = f_LO1 - f_LO2 + xffts_upper
    elif f_LO2 == 3.9:
        usb_lower = f_LO1 + f_LO2 + xffts_lower
        usb_upper = f_LO1 + f_LO2 + xffts_upper
        lsb_lower = f_LO1 - f_LO2 - xffts_upper
        lsb_upper = f_LO1 - f_LO2 - xffts_lower
    else:
        raise ValueError('f_LO2 must be eather 3.9 or 8.1')
        
    if ax is None:
        ax = plt.gca()

    # plot LSB and USB ranges
    style = {'lw': 0, 'alpha': 0.5, 'ymin': 0, 'ymax': 0.4}
    ax.axvspan(lsb_lower, lsb_upper, color='royalblue', **style)
    ax.axvspan(usb_lower, usb_upper, color='hotpink', **style)

    style = {'color': 'white', 'ha': 'center', 'va': 'center'}
    ax.annotate('LSB', ((lsb_lower+lsb_upper)/2, 0.2), **style)
    ax.annotate('USB', ((usb_lower+usb_upper)/2, 0.2), **style)
    
    style = {'color': 'gray', 'ha': 'center', 'va': 'center', 'rotation': 90}
    ax.annotate(lsb_lower, (lsb_lower, 0.2), **style)
    ax.annotate(lsb_upper, (lsb_upper, 0.2), **style)
    ax.annotate(usb_lower, (usb_lower, 0.2), **style)
    ax.annotate(usb_upper, (usb_upper, 0.2), **style)

    # plot bars of lines
    if lines is None:
        lines = {}

    lines.update({'f_LO1': f_LO1})

    style = {'color': '#333333', 'ha': 'center', 'va': 'center'}
    for name, freq in lines.items():
        ax.axvline(freq, ymin=0, ymax=0.8, color='gray')
        ax.annotate('{0}\n({1:.1f})'.format(name, freq), (freq, 0.9), **style)

    # preferences
    ax.set_xlim(lsb_lower-1, usb_upper+1)
    ax.set_yticklabels([])
    ax.set_xlabel('Observed frequency (GHz)')
```

## Samples

```python
fig, ax = plt.subplots()

f_LO1 = 140.0
f_LO2 = 8.1
lines = {'SO2': 134.00486000, 'CS (3-2)': 146.96902870}

plot(f_LO1, f_LO2, lines=lines)
```
